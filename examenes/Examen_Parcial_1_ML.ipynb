## Resultados Mundiales

```python
cols = confirmed_cases.keys()
confirmed = confirmed_cases.loc[:,cols[4]:cols[-1]]
deaths = death_cases.loc[:,cols[4]:cols[-1]]

dates = confirmed.keys()
world_cases = []
total_deaths = []

for i in dates:
    confirmed_sum = confirmed[i].sum()
    death_sum = deaths[i].sum()
    world_cases.append(confirmed_sum)
    total_deaths.append(death_sum)

days_since = np.array([i for i in range(len(dates))]).reshape(-1,1)
world_cases = np.array(world_cases).reshape(-1,1)
total_deaths = np.array(total_deaths).reshape(-1,1)
```

<font color = "red">Describe que es lo que sucede en el bloque de código anterior</font>

Se obtienen los datos correspondientes a los casos por fechas de confirmed_cases y death_cases para realizar los conteos, sumando todos los casos de cada dia y anexandolos a world_cases y total_deaths respectivamente.
De ahi se aplica un reshape para tener n-1 filas conformadas por valores individuales que representa la cantidad de casos y muertes totales por dia, mientras que days_since muestra la cantidad de dias desde el primer caso, en
el mismo formato.

```python
days_in_future = 10   
future_forecast = np.array([i for i in range(len(dates) + days_in_future)]).reshape(-1,1)
adjusted_dates = future_forecast[:-10] 
```

```python
start = '1/22/2020'
start_date = datetime.datetime.strptime(start, '%m/%d/%Y')
future_forecast_dates = []
for i in range(len(future_forecast)):
    future_forecast_dates.append((start_date + datetime.timedelta(days=i)).strftime('%m/%d/%Y'))
```

### Grados

```python
from sklearn.preprocessing import PolynomialFeatures
poli_deg3 = PolynomialFeatures(degree = 3)
poli_deg4 = PolynomialFeatures(degree = 4)
poli_deg5 = PolynomialFeatures(degree = 5)
```

# Confirmados

```python
from sklearn.linear_model import LinearRegression
```

```python
def Predictions(x_train, x_test, y_train, y_test, df, country, types):

  ############ Grado 3 ###############
  X_train_poli3 = poli_deg3.fit_transform(x_train)
  X_test_poli3 = poli_deg3.fit_transform(x_test)

  pr3 = LinearRegression()

  pr3.fit(X_train_poli3, y_train)

  Y_pred_pr3 = pr3.predict(X_test_poli3)

  pred3 = pr3.predict(poli_deg3.fit_transform(future_forecast))

  plt.scatter(x_test, y_test)
  plt.plot(x_test, Y_pred_pr3, color='red', linewidth=1)
  plt.show()


  print('Linear regression future predictions:')
  print(pred3[-10:])


############ Grado 4 ###############
  X_train_poli4 = poli_deg4.fit_transform(x_train)
  X_test_poli4 = poli_deg4.fit_transform(x_test)

  pr4 = LinearRegression()

  pr4.fit(X_train_poli4, y_train)

  Y_pred_pr4 = pr4.predict(X_test_poli4)

  pred4 = pr4.predict(poli_deg4.fit_transform(future_forecast))

  plt.scatter(x_test, y_test)
  plt.plot(x_test, Y_pred_pr4, color='blue', linewidth=1)
  plt.show()


  print('Linear regression future predictions:')
  print(pred4[-10:])


############ Grado 5 ###############
  X_train_poli5 = poli_deg5.fit_transform(x_train)
  X_test_poli5 = poli_deg5.fit_transform(x_test)

  pr5 = LinearRegression()

  pr5.fit(X_train_poli5, y_train)

  Y_pred_pr5 = pr5.predict(X_test_poli5)

  pred5 = pr5.predict(poli_deg5.fit_transform(future_forecast))

  plt.scatter(x_test, y_test)
  plt.plot(x_test, Y_pred_pr5, color='green', linewidth=1)
  plt.show()


  print('Linear regression future predictions:')
  print(pred5[-10:])

  plt.figure(figsize=(20,15))
  plt.plot(adjusted_dates, df)
  plt.plot(future_forecast, pred3, linestyle='dashed', color='red')
  plt.plot(future_forecast, pred4, linestyle='dashed', color='green')
  plt.plot(future_forecast, pred5, linestyle='dashed', color='brown')
  plt.title(country + ' Number of ' + types + ' over time', size=24)
  plt.xlabel('Days since 1/22/2020', size=24)
  plt.ylabel('Number of ' + types, size = 24)
  plt.legend(['Confirmed ' + types, 'LR predictions 3', 'LR predictions 4', 'LR predictions 5'])
  plt.xticks(size=12)
  plt.yticks(size=12)
  plt.show()
```

```python
########## LIBRERÍAS A UTILIZAR ##########
#Se importan la librerias a utilizar
from sklearn import datasets
from sklearn.model_selection import train_test_split

########## PREPARAR LA DATA ##########
#Importamos los datos de la misma librería de scikit-learn
dataset = datasets.load_breast_cancer()
```

```python
dataset
```

```python
dataset.keys()
```

```python
#Seleccionamos todas las columnas

#Defino los datos correspondientes a las etiquetas
X=dataset.data
y=dataset.target
```

```python
########## IMPLEMENTACIÓN DE K VECINOS MÁS CERCANOS ##########

#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
print("#----------------------------------#")
algDTree = DecisionTreeClassifier()
algDTree.fit(X_train, y_train)
y_pred = algDTree.predict(X_test)
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average='micro')
print('Precisión del modelo:')
print(precision)
```

```python
from sklearn.ensemble import RandomForestClassifier
print("#----------------------------------#")
algRFC = RandomForestClassifier()
algRFC.fit(X_train, y_train)
y_pred = algRFC.predict(X_test)
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average='micro')
print('Precisión del modelo:')
print(precision)
```

```python
from sklearn.svm import SVC
print("#----------------------------------#")
algSVC = SVC()
algSVC.fit(X_train, y_train)
y_pred = algSVC.predict(X_test)
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average='micro')
print('Precisión del modelo:')
print(precision)
```
